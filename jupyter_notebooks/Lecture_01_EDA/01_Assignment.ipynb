```python
#4-5. What are the mean and standard deviation of age for 
# those who earn more than 50K per year (salary feature) and those who earn less than 50K per year?
```

```python
more_std = df[df['salary'] == 51]['age'].std()
more_mean = df[df['salary'] == 51]['age'].mean()
print("More then 50K STD :", more_std)
print("More then 50K Mean :", more_mean)
```

```python
less_std = df[df['salary'] == 50]['age'].std()
less_mean = df[df['salary'] == 50]['age'].mean()
print("Less then 50K STD :", less_std)
print("Less then 50K Mean :", less_mean)
```
